```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.loadtxt('data.csv')

Ngrid = 100

limits = np.array([[0,0], [0,Ngrid], [Ngrid, Ngrid], [Ngrid,0], [0,0]])


fig, ax = plt.subplots(figsize=(4,4))

ax.scatter(np.floor(data/Ngrid), data%Ngrid, c='k', s=5)

plt.plot(limits[:,0], limits[:,1], 'k--')

plt.savefig('Images/300.png')

plt.show()
```
